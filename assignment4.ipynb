```python
import pandas as pd
import plotly.express as px

pd.set_option("display.max_rows", None)

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

df.shape
```

```python
df.sample(5)
```

```python
df.isna().sum()
```

```python
df[["Year", "Country","GDP per capita (current US$)", "Life expectancy at birth, total (years)"]].sample(5)
```

```python
df2 = df[df["Year"] != 2022]
df2.shape
```

```python
df2.isna().sum()
```

```python
df3 = df2.dropna()
df3.isna().sum()
```

```python
df3.shape
```

#Homework Question 1:

to check if this dataset has duplicate rows. If it has, only keep one from the duplicated rows and drop the rest.



```python
# Check for duplicate rows
duplicate_rows = df3.duplicated()
print(f"Number of duplicate rows: {duplicate_rows.sum()}")

# Drop duplicate rows, keeping the first occurrence
df4 = df3.drop_duplicates()

# Check the shape after dropping duplicates
print(f"Shape of the dataset after dropping duplicates: {df4.shape}")

# Optionally, verify if there are any remaining duplicates
remaining_duplicates = df4.duplicated().sum()
print(f"Remaining duplicates after dropping: {remaining_duplicates}")
```

#HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
df_year = df4[df4["Year"] == 2015]

fig = px.scatter(df_year,
                 x="GDP per capita (current US$)",
                 y="Life expectancy at birth, total (years)",
                 title="Relationship between Wealth (GDP per capita) and Health (Life Expectancy) in 2015",
                 labels={
                     "GDP per capita (current US$)": "GDP per capita (current US$)",
                     "Life expectancy at birth, total (years)": "Life Expectancy (years)"
                 },
                 hover_name="Country")

fig.show()
```

#Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
df_population_year = df4[df4["Year"] == 2015]

df_population_region = df_population_year.groupby("Region")["Population, total"].sum().reset_index()

fig_pie = px.pie(df_population_region,
                 values="Population, total",
                 names="Region",
                 title="Population Distribution by Region in 2015",
                 labels={
                     "Population, total": "Total Population",
                     "Region": "Region"
                 })

fig_pie.show()
```

#Home work question 4

remove the color legend.
Order the above bar chart based on the values of the average GDP per capita.

```python
df_group = df_group.sort_values(by="GDP per capita (current US$)", ascending=False)

fig = px.bar(df_group,
             x="Region",
             y="GDP per capita (current US$)",
             title="Average GDP per capita by Region in 2020 (Ordered)",
             labels={
                 "GDP per capita (current US$)": "Average GDP per capita (US$)",
                 "Region": "Region"
             })


fig.update_traces(showlegend=False)


fig.show()
```
