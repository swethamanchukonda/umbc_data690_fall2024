```python
import pandas as pd
import plotly.express as px

# Load the data
DATA_URL = "https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/fall2024/survey/UMBC%20DATA%20690%20FALL%202024%20DR%20WANG%20Student%20Survey.csv"
df = pd.read_csv(DATA_URL)

# Extracting and processing educational background data
df_edu_dist = df["1. My primary educational background"].value_counts().reset_index()
df_edu_dist.columns = ["Education Background", "Count"]

# Creating a pie chart to visualize the distribution of educational backgrounds
fig = px.pie(
    df_edu_dist,
    names="Education Background",
    values="Count",
    title="UMBC DATA 690 Fall 2024 Student Education Distribution",
    color_discrete_sequence=px.colors.qualitative.Set3
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the data
DATA_URL = "https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/fall2024/survey/UMBC%20DATA%20690%20FALL%202024%20DR%20WANG%20Student%20Survey.csv"
df = pd.read_csv(DATA_URL)

# Process the data: Split courses, clean, and count occurrences
courses = df["12. Other courses taking this semester"].dropna().str.split(",").explode().str.strip().str.lower()
course_counts = courses.value_counts().reset_index()
course_counts.columns = ["Course", "Count"]

# Create a bar chart
fig_bar = px.bar(course_counts, x='Course', y='Count', title='Distribution of Other Courses - Bar Chart')

# Show the bar chart
fig_bar.show()

# Create a pie chart
fig_pie = px.pie(course_counts, names='Course', values='Count', title='Distribution of Other Courses - Pie Chart')

# Show the pie chart
fig_pie.show()
```
