#Assignment 03 Part One

Remove the color legend in the above bar char
Add the number to the bar
Add text annotation for "Data source" and "Graph source"
Change the title for Y-axis from "count" to "Number of Students"

```python
import pandas as pd
import plotly.express as px

# Load the dataset
DATA_URL = "https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/fall2024/survey/UMBC%20DATA%20690%20FALL%202024%20DR%20WANG%20Student%20Survey.csv"
df = pd.read_csv(DATA_URL)

# Define the df_edu_dist DataFrame again
df_edu_dist = df["1. My primary educational background"].value_counts().reset_index()
df_edu_dist.columns = ['Educational Background', 'Count']

# Create the bar chart with the necessary modifications
fig = px.bar(
    df_edu_dist,
    x="Educational Background",
    y="Count",
    text="Count",  # Add the number to the bar
    title="UMBC DATA 690 Fall 2024 Student Education Distribution"
)

# Remove the color legend
fig.update_layout(showlegend=False)

# Add text annotations for "Data source" and "Graph source"
fig.add_annotation(
    text="Data source: UMBC DATA 690 Fall 2024 Survey",
    xref="paper", yref="paper",
    x=0, y=-0.2,
    showarrow=False
)

fig.add_annotation(
    text="Graph source: Generated by Student Code",
    xref="paper", yref="paper",
    x=1, y=-0.2,
    showarrow=False
)

# Change the Y-axis title
fig.update_yaxes(title_text="Number of Students")

# Show the chart
fig.show()
```

#Assignment 03 Part 2

Process column "11. What is your long-term career goal?" to find out the most popular career goals of the class. User Python code to parse the responses and count occurences of words, then use your naked eyes to make determination. Write your observation/analysis using Markdown.

```python
# Get the list of answers for the long-term career goals
career_goals_list = df["11. What is your long-term career goal?"].dropna().tolist()

# Initialize an empty dictionary to store word frequency
career_goal_word_freq = {}

# Process each answer
for answer in career_goals_list:
    # Split the answer into words
    list_words = answer.split()
    for word in list_words:
        # Update the word frequency count
        if word in career_goal_word_freq:
            career_goal_word_freq[word] += 1
        else:
            career_goal_word_freq[word] = 1

# Sort the word frequency dictionary by the count
sorted_career_goal_word_freq = dict(sorted(career_goal_word_freq.items(), key=lambda item: item[1], reverse=True))

# Print the sorted word frequency dictionary
sorted_career_goal_word_freq
```
