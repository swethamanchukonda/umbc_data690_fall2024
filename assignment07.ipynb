step 1 - create a column "Life Expectancy Quartile" similar to the GDP per Capita Quartile above. Step 2 - Create a column "health and wealth status" based on the following definition

rich and healthy (1st GDP per capita quartile and 1st life expentancy quartile)
poor and unhealthy ((4th GDP per capita quartile and 4th life expentancy quartile)
rich and unhealthy (1st GDP per capita quartile and 4th life expentancy quartile)
poor and healthy ((4th GDP per capita quartile and 1st life expentancy quartile)
Other


```python
import pandas as pd

# Load the dataset from the provided URL
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

# Rename columns for clarity
column_names_dict = {
    "GDP per capita (current US$)" : "GDP per Capita",
    "Life expectancy at birth, total (years)" : "Life Expectancy",
    "Population, total" : "Population"
}
df = df.rename(columns=column_names_dict)

# Step 1: Define GDP Quartile assignment function and create the "GDP Quartile" column
def assign_gdp_quartile(gdp):
    if gdp <= 1862.89:
        return "1st Quartile (<=25%)"
    elif gdp > 1862.89 and gdp <= 6048.30:
        return "2nd Quartile (25-50%)"
    elif gdp > 6048.30 and gdp <= 22137.31:
        return "3rd Quartile (50-75%)"
    else:
        return "4th Quartile (>75%)"

df["GDP Quartile"] = df["GDP per Capita"].apply(assign_gdp_quartile)

# Step 2: Create the "Life Expectancy Quartile" column
def assign_life_expectancy_quartile(life_expectancy):
    if life_expectancy <= 64.56:
        return "1st Quartile (<=25%)"
    elif life_expectancy > 64.56 and life_expectancy <= 71.96:
        return "2nd Quartile (25-50%)"
    elif life_expectancy > 71.96 and life_expectancy <= 77.85:
        return "3rd Quartile (50-75%)"
    else:
        return "4th Quartile (>75%)"

df["Life Expectancy Quartile"] = df["Life Expectancy"].apply(assign_life_expectancy_quartile)

# Step 3: Create the "Health and Wealth Status" column based on both GDP and Life Expectancy quartiles
def health_and_wealth_status(gdp_quartile, life_quartile):
    if gdp_quartile == "1st Quartile (<=25%)" and life_quartile == "1st Quartile (<=25%)":
        return "rich and healthy"
    elif gdp_quartile == "4th Quartile (>75%)" and life_quartile == "4th Quartile (>75%)":
        return "poor and unhealthy"
    elif gdp_quartile == "1st Quartile (<=25%)" and life_quartile == "4th Quartile (>75%)":
        return "rich and unhealthy"
    elif gdp_quartile == "4th Quartile (>75%)" and life_quartile == "1st Quartile (<=25%)":
        return "poor and healthy"
    else:
        return "Other"

# Apply the function to create the "Health and Wealth Status" column
df["Health and Wealth Status"] = df.apply(lambda row: health_and_wealth_status(row["GDP Quartile"], row["Life Expectancy Quartile"]), axis=1)

# Display sample data to verify
df[["GDP Quartile", "Life Expectancy Quartile", "Health and Wealth Status"]].head()
```
